```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df_gym = pd.read_csv("data/gym_members_exercise_tracking.csv")
```

```python
df_gym
```

```python
df_gym.head(20)
```

```python
df_gym.info()
```

```python
df_gym.describe()
```

```python
sns.countplot(data=df_gym, x="Gender")
plt.title("Count Of Gender")
```

```python
sns.histplot(data=df_gym, x='Age', kde=True)
plt.title("Distrbution Of Age")
```

```python
sns.barplot(data=df_gym, x='Workout_Type', y='Calories_Burned', hue='Gender')
plt.title("Calories Burned For Each Workout Type")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
new = df_gym[['Weight (kg)', 'Height (m)', 'BMI', 'Calories_Burned', 'Session_Duration (hours)', "Gender"]]
sns.pairplot(data=new, hue='Gender', palette='coolwarm')
```

```python
sns.scatterplot(data=df_gym, x='Session_Duration (hours)', y='Calories_Burned', hue='Experience_Level', color='set1')
plt.title("Claories Burned By Session Duration And Experience Level")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python

# Compute the correlation matrix
#corr = df_gym.corr()

# Plot heatmap
f, ax = plt.subplots(figsize=(13, 10))
sns.heatmap(df_gym.corr(),
            annot=True,
            linewidths=.5,
            ax=ax,
            vmin=-1,
            vmax=1);
```
