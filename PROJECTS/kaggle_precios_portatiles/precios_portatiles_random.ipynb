```python
import pandas as pd
import numpy as np

# Cargar el archivo sample_submission.csv
sample_submission = pd.read_csv('data/sample_submission.csv')

# Generar un precio aleatorio único para todas las predicciones, redondeado a 1 decimal
random_price = round(np.random.uniform(500, 3000), 1)  # Rango de precios entre 500 y 3000 euros, redondeado a 1 decimal

# Crear el DataFrame para las predicciones con un único precio aleatorio
submission_random_price = sample_submission.copy()
submission_random_price['Price_euros'] = random_price

# Guardar el archivo de salida como CSV
submission_random_price.to_csv('random_price_predictions.csv', index=False)
```
