## Building a regression model

The data have already been split to predictor (*data*) and response (*target*) variables. Given this information, we'll apply what we have previously learned about linear regression and apply the algorithm to the diabetes dataset.

#### In the cell below, import the `linear_model` class from `sklearn`. 

```python
# Importar la clase 'linear_model' desde 'sklearn'
from sklearn import linear_model
```

#### Create a new instance of the linear regression model and assign the new instance to the variable `diabetes_model`.

```python
from sklearn.linear_model import LinearRegression

# Crear una nueva instancia del modelo de regresión lineal
diabetes_model = LinearRegression()

# Mostrar el modelo para confirmar la creación
print("Instancia del modelo de regresión lineal creada:")
diabetes_model
```

#### Next, let's split the training and test data.

Define `diabetes_data_train`, `diabetes_target_train`, `diabetes_data_test`, and `diabetes_target_test`. Use the last 20 records for the test data and the rest for the training data.

```python
# Definir los datos de entrenamiento y prueba
# Usaremos los últimos 20 registros para el conjunto de prueba
diabetes_data_train = diabetes['data'][:-20]
diabetes_target_train = diabetes['target'][:-20]
diabetes_data_test = diabetes['data'][-20:]
diabetes_target_test = diabetes['target'][-20:]

# Mostrar las dimensiones de los conjuntos de datos
print(f"Dimensión de 'diabetes_data_train': {diabetes_data_train.shape}")
print(f"Dimensión de 'diabetes_target_train': {diabetes_target_train.shape}")
print(f"Dimensión de 'diabetes_data_test': {diabetes_data_test.shape}")
print(f"Dimensión de 'diabetes_target_test': {diabetes_target_test.shape}")
```

Perform a brief exploratory analysis to investigate the conditions necessary for linear regression. This analysis will help you understand the distribution of the target variable and its correlations with other features.

```python
# 1. Análisis de la distribución de la variable objetivo (target)
plt.figure(figsize=(8, 5))
sns.histplot(df['target'], kde=True, color='blue')
plt.title('Distribución de la Variable Objetivo (target)')
plt.xlabel('Progresión de la Enfermedad')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()

# 2. Análisis de correlación entre las características y la variable objetivo
correlation_matrix = df.corr()

# Mostrar el mapa de calor de la correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones')
plt.show()

# 3. Relación entre cada característica y la variable objetivo mediante gráficos de dispersión
plt.figure(figsize=(15, 10))
for i, feature in enumerate(diabetes['feature_names']):
    plt.subplot(3, 4, i + 1)
    sns.scatterplot(x=df[feature], y=df['target'], color='purple', edgecolor='w')
    plt.title(f'{feature} vs target')
    plt.xlabel(feature)
    plt.ylabel('target')
plt.tight_layout()
plt.show()
```

Fit the training data and target to `diabetes_model`. Print the *intercept* and *coefficients* of the model.

```python
# Ajustar los datos de entrenamiento al modelo de regresión lineal
diabetes_model.fit(diabetes_data_train, diabetes_target_train)

# Obtener la intersección (intercepto) y los coeficientes del modelo
intercept = diabetes_model.intercept_
coefficients = diabetes_model.coef_

# Imprimir la intersección y los coeficientes
print("Intersección (Intercepto):", intercept)
print("Coeficientes (Coefficients):", coefficients)

# Mostrar los coeficientes junto con los nombres de las características
feature_names = diabetes['feature_names']
print("\nCoeficientes del modelo:")
for feature, coef in zip(feature_names, coefficients):
    print(f"{feature}: {coef:.4f}")
```

#### Inspecting the results

From the outputs you should have seen:

- The intercept is a float number.
- The coefficients are an array containing 10 float numbers.

This is the linear regression model fitted to your training dataset.

#### Using your fitted linear regression model, predict the *y* of `diabetes_data_test`.

y = a + b*x1 + c*x2 + d*x3 + ....

```python
# Realizar predicciones utilizando el modelo de regresión lineal ajustado
y_pred = diabetes_model.predict(diabetes_data_test)

# Mostrar las predicciones realizadas
print("Predicciones del conjunto de prueba:")
print(y_pred)
```

#### Print your `diabetes_target_test` and compare with the prediction. 

```python
# Crear un DataFrame para comparar los valores reales y las predicciones del modelo
comparison_df = pd.DataFrame({
    'Actual': diabetes_target_test,
    'Predicted': y_pred
})

# Mostrar las primeras 10 filas de la comparación
print("Comparación entre valores reales y predicciones:")
print(comparison_df.head(10))
```

#### Is `diabetes_target_test` exactly the same as the model prediction? Use different error metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Calcular diferentes métricas de error
mae = mean_absolute_error(diabetes_target_test, y_pred)
mse = mean_squared_error(diabetes_target_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(diabetes_target_test, y_pred)

# Imprimir los resultados de las métricas de error
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R²): {r2:.4f}")
```

#### Iterate with a new model to assess if any improvements can be made compared to the previous one. Consider reducing the number of variables or using `StandardScaler` to enhance model performance.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Escalar los datos con StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(diabetes_data_train)
X_test_scaled = scaler.transform(diabetes_data_test)

# Seleccionar las mejores 5 características utilizando SelectKBest
selector = SelectKBest(score_func=f_regression, k=5)
X_train_selected = selector.fit_transform(X_train_scaled, diabetes_target_train)
X_test_selected = selector.transform(X_test_scaled)

# Crear un nuevo modelo de regresión lineal con las características seleccionadas
improved_model = LinearRegression()

# Ajustar el modelo con las características seleccionadas
improved_model.fit(X_train_selected, diabetes_target_train)

# Realizar predicciones con el nuevo modelo mejorado
y_pred_improved = improved_model.predict(X_test_selected)

# Evaluar el nuevo modelo utilizando métricas de error
mae_improved = mean_absolute_error(diabetes_target_test, y_pred_improved)
mse_improved = mean_squared_error(diabetes_target_test, y_pred_improved)
rmse_improved = np.sqrt(mse_improved)
r2_improved = r2_score(diabetes_target_test, y_pred_improved)

# Imprimir los resultados del nuevo modelo mejorado
print("Resultados del modelo mejorado utilizando StandardScaler y selección de características:")
print(f"Mean Absolute Error (MAE): {mae_improved:.4f}")
print(f"Mean Squared Error (MSE): {mse_improved:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_improved:.4f}")
print(f"R-squared (R²): {r2_improved:.4f}")

# Mostrar las características seleccionadas
selected_features = [feature for feature, selected in zip(diabetes['feature_names'], selector.get_support()) if selected]
print("\nCaracterísticas seleccionadas:", selected_features)
```
