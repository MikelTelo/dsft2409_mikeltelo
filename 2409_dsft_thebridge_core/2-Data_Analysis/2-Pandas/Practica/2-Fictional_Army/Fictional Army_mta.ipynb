### Step 1. This is the data given as a dictionary. Create a dataframe and assign it to a variable called army.

### Step 3. Drop features, if there are more than 30% missing values.

```python
#calcular el porcentaje de los valores faltantes
missing_percentage = army.isnull().mean() * 100
missing_percentage
```

```python
#calcular si hay mas de un 30% faltante
columns_to_drop = missing_percentage[missing_percentage > 30].index
columns_to_drop
```

```python
#eliminar la coljumna con mas del 30% nan
army_cleaned = army.drop(columns='readiness')
army_cleaned
```

### Step 4. Fill missing values with the mean of their regiment.

```python
#Calcular la media de deaths por regimiento

medias_por_grupo = army_cleaned.groupby('regiment')['deaths'].mean()
medias_por_grupo

```
