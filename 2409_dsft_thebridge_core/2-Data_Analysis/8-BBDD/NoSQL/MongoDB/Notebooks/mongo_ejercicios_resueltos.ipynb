```python
# https://stackoverflow.com/questions/20175122/how-can-i-use-not-like-operator-in-mongodb

where = {
        "cuisine": {"$not":{"$regex": "American "}},

        "grades.score": {"$gt": 70},
        
        "address.coord.1": {"$lt": -65.754168}}

select =  {"_id":0, "cuisine":1, "grades":1, "address":1, "name":1}

for x in restaurantes.find(where, select):
    print(x)
    print("-"*40)
```

```python
# 7

where = {
        "cuisine": {"$not":{"$regex": "American "}},
        "grades.grade": "A",
        "borough": {"$not":{"$regex": "Brooklyn"}},}

select =  {"_id":0, "cuisine":1, "grades":1, "name":1, "borough":1}

for x in restaurantes.find(where, select).sort("cuisine", -1):
    print(x)
    print("-"*40)
```

```python
# 8
# https://www.mongodb.com/docs/manual/reference/operator/query/or/

where = { "borough": "Bronx" , 
            "$or" : [{ "cuisine" : "American " },{ "cuisine" : "Chinese" } ] } 
select =  {"_id":0, "cuisine":1,"name":1, "borough":1}

for x in restaurantes.find(where, select):
    print(x)
    print("-"*40)
```

```python
# 9

# https://www.mongodb.com/docs/manual/reference/operator/query/in/

where = { "borough": {"$in" : ["Staten Island","Queens","Bronx","Brooklyn" ] } }
select =  {"_id":0, "cuisine":1,"name":1, "borough":1, "restaurant_id":1}

for x in restaurantes.find(where, select):
    print(x)
    print("-"*40)
```
