7. Cuánto ha vendido cada empleado

```python
query = '''
SELECT 
    e.FirstName ||' ' ||  e.LastName as Empleado,
    COUNT(DISTINCT i.InvoiceId) as Numero_Facturas,
    COUNT(DISTINCT c.CustomerId) as Numero_Clientes,
    ROUND(SUM(i.Total), 2) as Total_Ventas
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent'
GROUP BY e.EmployeeId
ORDER BY Total_Ventas DESC;
'''
crsr.execute(query)

# Obtener los resultados
results = crsr.fetchall()
# Obtener los nombres de las columnas
column_names = [description[0] for description in crsr.description]
# Crear una tabla formateada
table = tabulate(results, headers=column_names, tablefmt="pretty")
# Mostrar la tabla (esto funcionará automáticamente en un entorno interactivo)
print(table)
```

8. ¿Quién ha sido el agente de ventas que más ha vendido en 2009?

```python
query = '''
 SELECT 
    e.FirstName ||' ' ||  e.LastName as Empleado,
    COUNT(DISTINCT i.InvoiceId) as Numero_Facturas,
    COUNT(DISTINCT c.CustomerId) as Numero_Clientes,
    ROUND(SUM(i.Total), 2) as Total_Ventas
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE e.Title = 'Sales Support Agent' 
    AND strftime('%Y', i.InvoiceDate) = '2009'
GROUP BY e.EmployeeId
ORDER BY Total_Ventas DESC
LIMIT 1 
'''
crsr.execute(query)

# Obtener los resultados
results = crsr.fetchall()
# Obtener los nombres de las columnas
column_names = [description[0] for description in crsr.description]
# Crear una tabla formateada
table = tabulate(results, headers=column_names, tablefmt="pretty")
# Mostrar la tabla (esto funcionará automáticamente en un entorno interactivo)
print(table)
```

9. ¿Cuáles son los 3 grupos que más han vendido?

```python
query = '''
SELECT 
        ar.Name,
        COUNT(il.TrackId) as canciones_vendidas,
        COUNT(DISTINCT i.InvoiceId) as num_facturas,
        ROUND(SUM(il.UnitPrice * il.Quantity), 2) as total_ventas,
        GROUP_CONCAT(DISTINCT al.Title) as albumes
    FROM artists ar
    JOIN albums al ON ar.ArtistId = al.ArtistId
    JOIN tracks t ON al.AlbumId = t.AlbumId
    JOIN invoice_items il ON t.TrackId = il.TrackId
    JOIN invoices i ON il.InvoiceId = i.InvoiceId
    GROUP BY ar.ArtistId
    ORDER BY total_ventas DESC
    LIMIT 3
'''
crsr.execute(query)

# Obtener los resultados
results = crsr.fetchall()
# Obtener los nombres de las columnas
column_names = [description[0] for description in crsr.description]
# Crear una tabla formateada
table = tabulate(results, headers=column_names, tablefmt="pretty")
# Mostrar la tabla (esto funcionará automáticamente en un entorno interactivo)
print(table)
```
